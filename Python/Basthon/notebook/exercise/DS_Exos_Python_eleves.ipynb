### Exo 1 : score du rugby **
Ecrire un programme pour compter le score au rugby. On donnera une chaine de caractères correspondant aux différents types de points marqués : <br>
- e = essai non transformé = 5 points ;
- t = essai transformé = 7 points ; 
- p = pénalité = 3 point.
<br>
<br>
Exemple : si on rentre la chaine de caractère 'eepp' le programme retournera un score de 16 (soit 5 + 5 + 3 + 3)


```python
score_debut = input("Entrez les initiales en minuscule des types de points marqués:\n e, t, p\n")
score_final = 0

for i in score_debut:
    if i == "e":
        score_final += 5
    if i == "t":
        score_final += 7
    if i == "p":
        score_final += 3
print(score_final)
```
